```python
import sys
import subprocess
import pkg_resources

required = {'autoviz'}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed

if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)

import pandas as pd
import numpy as np
from autoviz import AutoViz_Class
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
students = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/students.csv")
students.head()
```

##### If you don’t have a target variable you can leave it blank:

```python
AV = AutoViz_Class()

%matplotlib inline

# Generate visualizations
AV.AutoViz("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/students.csv")
```

```python
students.sort_values(by='Salary', ascending=True, inplace=True)
```

```python
AV = AutoViz_Class()
sep = ','


dft = AV.AutoViz(students, sep, depVar="Salary", dfte="",
           header=0, verbose=1,
            lowess=False,chart_format='bokeh',max_rows_analyzed=500,max_cols_analyzed=30)
```
