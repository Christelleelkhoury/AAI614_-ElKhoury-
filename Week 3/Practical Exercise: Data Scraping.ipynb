```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

html = urlopen('https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html')
bs = BeautifulSoup(html, 'html.parser')
bs
```

```python
bs.find_all('table')
```

```python
table2 = bs.find_all('table')[1]
col_titles = table2.find_all('th')
col_titles
```

```python
table2_col_titles = [title.text for title in col_titles]
print(table2_col_titles)
```

```python
#bs.find('table', class_ ='table table-condensed table-striped table-hover table-bordered small')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table2_col_titles)
df
```

```python
column_data = table2.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  table2_rows = [data.text for data in row_data]
  #lenghth = len(df)
  #df.loc[lenghth] = table2_rows
  print(table2_rows)
```
